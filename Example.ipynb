```python
# Initialization
import pandas as pd
from Vaccine import VaccineStudy
import utilities
```

```python
# Choosing parameters

n_per_arm = 50
#sizes = [4,5,6,7,8] # the nhouseholds are divided evenly among the sizes
sizes = [8]

hsar = 0.2 # the desired household sar when averaged over the sizes of the households
hsarv = 0.01

es = 0.7 # Would we prefer another name for this?

## 

duration = 90 # days
importation_probability_over_study = 0.03
importation_rate = 1-(1-importation_probability_over_study)**(1/duration) # converting risk over study period to daily risk

from Vaccine import vaccinate_one # vaccinates one individual in each household
vaccination_method = vaccinate_one
```

```python
one_vax_study = VaccineStudy("one vaccinated", n_per_arm, sizes, hsar, es, (hsarv, 'hsarv'), duration, importation_rate, vaccination_method)
```

```python
# Running the study some number of times
trials = 2
vax_df, control_df = one_vax_study.run_trials(trials)
utilities.make_bar_chart(pd.concat([vax_df, control_df]))
```
